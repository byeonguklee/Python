```python
import bs4      # Crawling 역할
```

```python
import requests # 통신 역할
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text # requests가 html을 text로 가져옴
result
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser') # parser: 해석기
```

```python
bsObj.find('a', {'class':'basicList_link__JLQJf'}).text 
# a tag의 'basicList_link__JLQJf' class 찾기
```

```python
items = bsObj.find_all('a', {'class':'basicList_link__JLQJf'}) 
# a tag의 'basicList_link__JLQJf' class 모두! 찾기 -> 찾은 후 list로 표현
```

```python
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```
