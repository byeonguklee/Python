```python
fig = plt.figure()
ax = plt.axes()  # 축을 먼저 만들고

x = np.linspace(0,10,100)
ax.plot(x,np.sin(x)) #만든 축에 그래프 그리기
```

```python
plt.plot(x,np.sin(x-0),color='red')         # 0 ~ 10 np.linspace 범위가 줄어듬 // -1 ~ 9 ...
plt.plot(x,np.sin(x-1),color='chartreuse')  # 색 고유의 이름
plt.plot(x,np.sin(x-2),color='g')           # 짧은 색상코드 (g=green, K=black), RGB, CMYK
plt.plot(x,np.sin(x-3),color='0.75')        # 0 ~ 1사이에 회색조를 표현한다. 1:흰색 0:검은색 (빛의 밝기)
plt.plot(x,np.sin(x-4),color='#FFDD44')     # 16진수 RGB Color
plt.plot(x,np.sin(x-5),color=(1.0,0.2,0.3)) # RGB 값을 10진법의 튜플로 표현
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g') #solid '-' + green 'g' = '-g'
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

```python
#플롯 조정하기: 축 경계

plt.plot(x, np.sin(x))

plt.xlim(-1,11)    # 그려지는 그래프의 x축 범위
plt.ylim(-1.5,1.5) # 그려지는 그래프의 y축 범위
```

```python

plt.plot(x, np.sin(x))

plt.xlim(11,-1)    # 그려지는 그래프의 x축 방향 바꾸기 (좌우반전)
plt.ylim(1.5,-1.5) # 그려지는 그래프의 y축 방향 바꾸기 (상하반전)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5]) # 한번에 가능(앞에 두자리 x 축, 뒤에 두자리 y 축)
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
#플롯에 레이블 붙이기
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') #label: 선의 이름
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')                            #비율 동일하게
plt.legend()                                 #범주 넣기
```

```python
# 산점도의 출력

x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black') # 'o' 자리를 marker 라고함
```

```python
rng = np.random.RandomState(0) #임의의 수
for marker in ['o', ',', 'x', '+', 'v', '<', '>', 's','d','p']:
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
  plt.legend()
  plt.xlim(0,1.8)
```

```python
plt.plot(x,y,'-ok') # '-': solid, 'o':산점도, 'k':black
```

```python
plt.plot(x, y, '-p', color='gray', markersize=15, 
         linewidth=4, markerfacecolor='white',
         markeredgecolor='green', markeredgewidth='2')
```
